**¿Qué es GitHub?**
GitHub es una plataforma de alojamiento de código en la nube que utiliza Git, un sistema de control de versiones distribuido. En esencia, actúa como un gran centro social y de colaboración para proyectos de software.

Conceptos Clave
**Git:** El sistema de control de versiones. Es el software que rastrea y gestiona los cambios en tu código a lo largo del tiempo, permitiéndote volver a versiones anteriores, trabajar en paralelo, etc.

**GitHub:** La plataforma web que aloja los repositorios de Git, permitiendo la colaboración en equipo, la gestión de proyectos y la visibilidad de código (tanto público como privado).

**Repositorio (Repo):** Es la carpeta principal de tu proyecto. Contiene todos los archivos del proyecto, así como el historial completo de revisiones de Git.

**Commit (Confirmación):** Una instantánea de los cambios que has guardado en el historial de tu proyecto. Cada commit incluye un mensaje descriptivo.

**Branch (Rama):** Una línea de desarrollo independiente. Se usan para trabajar en nuevas características o corregir errores sin afectar la versión principal del proyecto (main o master).

**Comandos Esenciales de Git**
Los comandos que utilizas para interactuar con GitHub se ejecutan a través de la interfaz de línea de comandos de Git (Git Bash, Terminal, PowerShell, etc.) y se conocen como comandos de Git.

1. **Configuración Inicial y Repositorios**
Comando	Función	Ejemplo
`git config`	Configura tu nombre de usuario y correo electrónico globales para que se asocien a tus commits.
    `git config --global user.name "Tu Nombre"`

`git init`	Inicializa un nuevo repositorio de Git en la carpeta local actual.	
    `git init`

`git clone`	Descarga una copia completa de un repositorio remoto (de GitHub, por ejemplo) a tu máquina local.	
    `git clone [URL_del_repo]`

`git remote add`	Vincula tu repositorio local con uno remoto en GitHub (se suele usar origin como nombre).
    `git remote add origin [URL_del_repo]`

2. **Guardar Cambios (El Flujo Básico)**
Este es el proceso fundamental para registrar y subir tus modificaciones: **Status** → **Add** → **Commit** → **Push**.

Comando	Función	Ejemplo
`git status`	Muestra el estado de tu directorio de trabajo (qué archivos han sido modificados, cuáles están en staging, etc.).	
    `git status`
`git add`	Prepara los archivos modificados para ser incluidos en el próximo commit (área de staging).	
    `git add nombre_archivo.js git add.` (para todos los cambios)
`git commit`	Guarda la instantánea de los cambios preparados en el historial local, junto con un mensaje descriptivo.	
    `git commit -m "Mensaje descriptivo del cambio"`
`git push`	Sube los commits locales al repositorio remoto (GitHub).
    `git push origin main`

3. **Ramas y Colaboración**
Las ramas son cruciales para el desarrollo en equipo y la gestión de características.

Comando	Función	Ejemplo
`git branch`	Lista todas las ramas locales o crea una nueva rama.	
    `git branch git branch nueva-feature`
`git checkout`	Te permite cambiar a una rama existente o, con -b, crear y cambiar a una nueva rama de inmediato.	
    `git checkout nombre-rama git checkout -b nueva-feature`
`git pull`	Descarga los cambios del repositorio remoto y los fusiona en tu rama local actual. Es bueno hacerlo antes de empezar a trabajar.	
    `git pull origin main`
`git merge`	Combina los cambios de una rama en la rama actual. Usado comúnmente para integrar una feature terminada en main.	
    `git merge nueva-feature`
`git log`	Muestra el historial de commits de tu repositorio.	
    `git log --oneline`
